```python
import fitz  # PyMuPDF
import os
from google.colab import files

def pdf_to_markdown(pdf_path, output_md="output.md", images_dir="images"):
    # Create directories
    os.makedirs(images_dir, exist_ok=True)

    doc = fitz.open(pdf_path)
    md_content = ""

    for page_num, page in enumerate(doc):
        # Extract text
        text = page.get_text("text")
        md_content += f"{text}\n\n"

        # Extract images
        img_list = page.get_images(full=True)
        for img_idx, img in enumerate(img_list):
            xref = img[0]
            base_image = doc.extract_image(xref)
            image_data = base_image["image"]
            image_ext = base_image["ext"]

            # Save image
            img_name = f"page_{page_num+1}_img_{img_idx+1}.{image_ext}"
            img_path = os.path.join(images_dir, img_name)
            with open(img_path, "wb") as img_file:
                img_file.write(image_data)

            # Add Markdown image reference
            md_content += f"![Image {img_idx+1}]({img_path})\n\n"

    # Save Markdown
    with open(output_md, "w") as md_file:
        md_file.write(md_content)

    return output_md, images_dir

# Usage
uploaded = files.upload()
pdf_name = list(uploaded.keys())[0]
output_md, images_dir = pdf_to_markdown(pdf_name)

# Download results
files.download(output_md)
!zip -r images.zip {images_dir}
files.download("images.zip")
```
